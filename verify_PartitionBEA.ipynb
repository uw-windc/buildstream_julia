```julia
names = Dict(
    :md0 => :md_0,
    :s0 => :s_0,
    :m0 => :m_0,
    :trn0 => :trn_0,
    :tm0 => :tm_0,
    :ta0 => :ta_0,
    :y0 => :y_0,
    :va0 => :va_0,
    :mrg0 => :mrg_0,
    :ys0 => :ys_0,
    :id0 => :id_0,
    :x0 => :x_0,
    :tax0 => :tax_0,
    :ms0 => :ms_0,
    :fs0 => :fs_0,
    :fd0 => :fd_0,
    :duty0 => :duty_0,
    :ts0 => :ts_0,
    :sbd0 => :sbd_0,
    :a0 => :a_0
);
```

```julia
wrong = Dict()

key = :y0

#include("PartitionBEA.jl");

for key in keys(old_GU.parameters)

    old_parm = old_GU[key]
    parm = GU[names[key]]

    sets = [GU[e] for e in parm.sets]
    for idx in Iterators.product(sets...)
        ind = [[e] for e in idx]
        if old_parm[ind...] != parm[ind...]
            if key ∈ keys(wrong)
                push!(wrong[key],idx)
            else
                wrong[key] = [idx]
            end
        end
    end
end
wrong
```
